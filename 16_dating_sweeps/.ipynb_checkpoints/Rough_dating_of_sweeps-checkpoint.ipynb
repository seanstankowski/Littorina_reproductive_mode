# Very rough dating of sweeps

```r
### read in data and packages
data <-read.csv("pi_pvt.csv",header=TRUE)
options(repr.matrix.max.cols=150) ## see all columns in head
head(data,88)
library(dplyr)
```

```r
## Create a function for calculating time from pi and mu (3.00e-08)

Time <- function(pi) {
  print(pi/3.00e-08)
}

Time(0.003)
```

```r
plot(data$pi_Egglayer, data$pi_Brooder_pvt_corrtected, xlim=c(0,0.012), ylim=c(0,0.012))
abline(0,1)

mean_pi_brooder<-mean(data$pi_Brooder_pvt_corrtected)
median_pi_brooder<-median(data$pi_Brooder_pvt_corrtected)
Time(mean_pi_brooder)
Time(median_pi_brooder)
```

```r
hist(data$pi_Brooder_pvt_corrtected,breaks = 50, xlim=c(0,0.015))
```

```r
pi <- (data$pi_Brooder_pvt_corrtected)
times <- as.numeric(lapply(pi, Time))
hist(times, breaks=30, xlim=c(250000,0))
ave_time <- mean(times)
ave_time
sd(times)
```
