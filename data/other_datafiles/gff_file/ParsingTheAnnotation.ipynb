```r
setwd("/Users/sstankow/Documents/GitHub/Littorina_reproductive_mode/data/other_datafiles/gff_file")
full.list <- read.csv('modified_gff.csv',header=TRUE)
DE.Tr <- read.csv('Erica_Selection_and_DiffExpression_subset.csv',header=TRUE)
merged <- merge(DE.Tr, full.list, all.X=TRUE)
library('dplyr')
```

```r
## calculate the length of each region
merged <- merged %>% mutate(length = end - start)
```

```r
## make a bed file
bed <- cbind(merged$seqid,merged$start,merged$end)
write.table(bed,"bedfile.bed", row.names=FALSE, col.names = FALSE, sep="\t", quote = FALSE)
```
